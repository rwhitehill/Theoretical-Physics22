```python
import matplotlib
matplotlib.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
matplotlib.rc('text', usetex = True)
from matplotlib import pyplot as plt

import numpy as np
```

```python
fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(7,5))

x = np.linspace(np.pi/100,99*np.pi/100,1000)
ax.plot(180*x/np.pi,np.sin(x/2)**-4,'k')

ax.set_xlabel(r'$\theta~({\rm degrees})$',size=20)
ax.set_ylabel(r'$\sin^{-4}(\theta/2)$',size=20)
ax.tick_params(which='major',axis='both',direction='in',labelsize=20)
ax.semilogy()

fig.savefig('Rutherford_plot.pdf',bbox_inches='tight')
plt.show()
```
